```python
#Resolução TPC4

import random

lista = []

def criar_lista():
    global lista
    tamanho = int(input("Digite o tamanho da lista: "))
    lista = [random.randint(1, 100) for _ in range(tamanho)]
    print("Lista criada com sucesso!")

def ler_lista():
    global lista
    lista = [int(x) for x in input("Digite os números da lista separados por espaço: ").split()]
    print("Lista lida com sucesso!")

def soma():
    soma = 0
    for n in lista:
        soma = soma + n
    print(f"A soma dos elementos da lista é: {soma}")

def media():
    global lista
    result = sum(lista) / len(lista)
    print(f"A média dos elementos da lista é: {result}")

def maior():
    global lista
    maior = lista[0] 
    for n in lista:
        if n > maior:
            maior = n
    print(f"O maior elemento da lista é: {maior}")

def menor():
    global lista
    menor = lista [0]
    for n in lista [1:]:
        if n < menor:
            menor = n
    print(f"O menor elemento da lista é: {menor}")

def esta_ordenada_crescente():
    global lista
    if all(lista[i] <= lista[i+1] for i in range(len(lista)-1)):
        print("A lista está ordenada por ordem crescente.")
    else:
        print("A lista não está ordenada por ordem crescente.")

def esta_ordenada_decrescente():
    global lista
    if all(lista[i] >= lista[i+1] for i in range(len(lista)-1)):
        print("A lista está ordenada por ordem decrescente.")
    else:
        print("A lista não está ordenada por ordem decrescente.")

def procura_elemento():
    global lista
    elemento = int(input("Digite o elemento que deseja procurar na lista: "))
    if elemento in lista:
        posicao = lista.index(elemento)
        print(f"O elemento {elemento} foi encontrado na posição {posicao}.")
    else:
        print(f"O elemento {elemento} não foi encontrado na lista.")

while True:
    print("\nMenu:")
    print("(1) Criar Lista")
    print("(2) Ler Lista")
    print("(3) Soma")
    print("(4) Média")
    print("(5) Maior")
    print("(6) Menor")
    print("(7) Esta Ordenada por ordem crescente")
    print("(8) Esta Ordenada por ordem decrescente")
    print("(9) Procura um elemento")
    print("(0) Sair")

    opcao = input("Escolha uma opção: ")
    
    if opcao == '1':
        criar_lista()
    elif opcao == '2':
        ler_lista()
    elif opcao == '3':
        soma()
    elif opcao == '4':
        media()
    elif opcao == '5':
        maior()
    elif opcao == '6':
        menor()
    elif opcao == '7':
        esta_ordenada_crescente()
    elif opcao == '8':
        esta_ordenada_decrescente()
    elif opcao == '9':
        procura_elemento()
    elif opcao == '0':
        print(f"Lista atual: {lista}")
        break
    else:
        print("Opção inválida. Tente novamente.")


      
```
